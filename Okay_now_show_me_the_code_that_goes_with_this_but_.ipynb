```
import numpy as np
from scipy.fftpack import fft, ifft

def simulate_soliton_retrieval(initial_data_phase):
    # Parameters for the Immortal Geometry
    N = 1024  # Spatial grid points
    L = 40.0  # Physical domain scale
    x = np.linspace(-L/2, L/2, N)
    dt = 0.01 # Time step
    steps = 500

    # The 252 Constant (Wave Stability Factor)
    gamma = 252 / 100.0

    # 1. Initialize the Probe Soliton (The "Key")
    # A sech-pulse representing the retrieval request
    probe = 1.0 / np.cosh(x + 10)

    # 2. Initialize the Stored Data (The "File")
    # A stationary soliton packet at the center of the medium
    # The phase (e.g., np.pi/4) encodes the information.
    data_packet = 1.0 / np.cosh(x) * np.exp(1j * initial_data_phase)

    # Combined Wave Field
    psi = probe + data_packet

    # Spectral k-space for dispersion calculations
    dk = 2 * np.pi / L
    k = np.fft.fftfreq(N, d=L/N) * 2 * np.pi

    # Time Evolution (Split-Step Fourier Method)
    for _ in range(steps):
        # Linear Step (Dispersion in the Abyss)
        psi = ifft(np.exp(-0.5j * k**2 * dt) * fft(psi))

        # Non-linear Step (Self-Phase Modulation)
        psi = np.exp(1j * gamma * np.abs(psi)**2 * dt) * psi

    return x, np.abs(psi)**2, np.angle(psi)

# Execution of the system (This line will now cause an error, but will be fixed by subsequent cells)
# coordinates, intensity, phase = simulate_soliton_retrieval()
```
