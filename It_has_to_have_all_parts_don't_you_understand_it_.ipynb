````
import io
import numpy as np
from PIL import Image
import ipywidgets as widgets
from google.colab import files
from IPython.display import display, clear_output, HTML # Import HTML for text formatting

# --- MODULE 2: CORE GEOMETRY LOGIC ---
# "When it's larger, the scale is smaller."
def transform_to_immortal_bitmap(image_data):
    img = Image.open(io.BytesIO(image_data)).convert('RGBA')
    img_array = np.array(img)

    # Apply the 252 filter for "immortal binary math"
    filtered_array = apply_252_filter(img_array)

    # Convert back to PIL Image, preserving alpha
    return Image.fromarray(filtered_array, 'RGBA')

def apply_252_filter(image_array):
    print(f"Shape of image_array in apply_252_filter: {image_array.shape}") # Added print statement
    # Mathematical sealer for the neural network
    # 252 acts as the anchor for the immortal geometry
    transformed_array = np.copy(image_array)
    # Apply a 'binary-style' thresholding based on the constant 252
    # For the RGB channels (first 3), set pixels to 0 or 255 based on a threshold
    threshold_value = 252 // 2 # Using 126 as a simple threshold for a binary-like effect
    transformed_array[:, :, :3] = (transformed_array[:, :, :3] > threshold_value) * 255

    return transformed_array

# Helper function to convert PIL Image to bytes for ipywidgets.Image
def pil_to_bytes(img):
    img_byte_arr = io.BytesIO()
    img.save(img_byte_arr, format='PNG')
    return img_byte_arr.getvalue()

# --- MODULE 3: INTERACTIVE COMPONENTS ---
class ImmortalNotebook:
    def __init__(self):
        self.layers = [] # Change to a list to hold multiple images
        self.upload_widget = widgets.FileUpload(accept='image/*', multiple=True)
        self.output_area = widgets.Output()
        self.save_button = widgets.Button(description="Export Blended Image")
        self.last_blended_image = None # To store the last blended image
        self.upload_widget.observe(self.process_uploads, names='value')
        self.save_button.on_click(self.on_save_button_clicked)

        # Initialize widgets that will be used in render_viewer and stored as instance variables
        self._slider = widgets.FloatSlider(
            value=0.5, min=0, max=1.0, step=0.01,
            description='Decryption:', continuous_update=True
        )
        self._toggle_buttons = widgets.ToggleButtons(
            options=['Blended Decryption', 'Base Image', 'Overlay Composite'], # Reordered for default
            description='View:',
            disabled=False,
            button_style='',
            tooltips=['Show the blended result using the slider', 'Show only the base image', 'Show only the composite overlay'],
            value='Blended Decryption'
        )
        self._image_display_output = widgets.Output() # Dedicated output for the main toggled image

    def process_uploads(self, change):
        with self.output_area:
            clear_output()
            file_data_dict = change['new']
            file_keys = list(file_data_dict.keys())

            if not file_keys:
                print("Error: Please upload at least one image.")
                self.layers = [] # Ensure layers is empty if nothing uploaded
                self.last_blended_image = None # Clear previous blended image
                return

            self.layers = [] # Clear previous layers
            for key in file_keys:
                self.layers.append(transform_to_immortal_bitmap(file_data_dict[key]['content']))

            print(f"Systems Online. Loaded {len(self.layers)} images.")
            # If only one image, display it immediately, otherwise render the viewer
            if len(self.layers) == 1:
                display(self.layers[0])
                print(f"Single image loaded. {file_keys[0]}")
                self.last_blended_image = None # No blend if only one image
            else:
                self.render_viewer()

    def on_save_button_clicked(self, b):
        with self.output_area:
            if self.last_blended_image:
                img_byte_arr = io.BytesIO()
                self.last_blended_image.save(img_byte_arr, format='PNG')
                img_byte_arr.seek(0)
                files.download(filename='blended_image.png', data=img_byte_arr.read())
                print("Blended image exported as blended_image.png")
            else:
                print("No blended image to export. Please ensure you have uploaded at least two images and adjusted the slider.")

    def render_viewer(self):
        # Helper function to update the main image display based on toggle selection
        def _update_main_image_display():
            if not self.layers or len(self.layers) < 2:
                return # Should not happen if render_viewer is called correctly

            base_image = self.layers[0]
            first_image_size = base_image.size

            overlay_composite = self.layers[1].resize(first_image_size)
            for i in range(2, len(self.layers)):
                next_layer = self.layers[i].resize(first_image_size)
                overlay_composite = Image.blend(overlay_composite, next_layer, alpha=0.5)

            blended = Image.blend(base_image, overlay_composite, alpha=self._slider.value)
            self.last_blended_image = blended # Store for save button

            with self._image_display_output: # Update only this specific output widget
                clear_output(wait=True)
                current_view = self._toggle_buttons.value
                if current_view == 'Base Image':
                    display(base_image)
                elif current_view == 'Overlay Composite':
                    display(overlay_composite)
                elif current_view == 'Blended Decryption':
                    display(blended)

        # Slider change handler: Clears main output area, re-displays controls, and updates main image.
        def on_slider_value_change(change):
            with self.output_area:
                clear_output(wait=True) # Clear entire main output area for other info
                if not self.layers:
                    print("No images to display.")
                    self.last_blended_image = None
                    return
                elif len(self.layers) == 1:
                    display(self.layers[0])
                    print(f"Only one image loaded. Slider has no effect. Key: 252")
                    self.last_blended_image = None
                    return

                # Display individual 'decrypted' bitmaps
                display(HTML('<b>Individual Decrypted Bitmaps (Base | Composite Overlay):</b>'))
                base_widget_image = widgets.Image(value=pil_to_bytes(self.layers[0]), format='png', width=self.layers[0].size[0], height=self.layers[0].size[1])
                # Recompute overlay_composite here for consistency with _update_main_image_display
                first_image_size = self.layers[0].size
                overlay_composite = self.layers[1].resize(first_image_size)
                for i in range(2, len(self.layers)):
                    next_layer = self.layers[i].resize(first_image_size)
                    overlay_composite = Image.blend(overlay_composite, next_layer, alpha=0.5)
                overlay_widget_image = widgets.Image(value=pil_to_bytes(overlay_composite), format='png', width=first_image_size[0], height=first_image_size[1])
                display(widgets.HBox([base_widget_image, overlay_widget_image]))

                display(HTML('<b>Final Output View:</b>'))
                display(self._slider)
                display(self._toggle_buttons)
                display(self._image_display_output) # Display the dedicated output container
                display(self.save_button)
                print(f"Current Polarity: {int(change['new']*100)}% | Key: 252")

                # Now, update the actual image displayed within _image_display_output
                _update_main_image_display()

        # Toggle button change handler: Only updates the image displayed in the dedicated output.
        def on_toggle_button_change(change):
            _update_main_image_display()

        # Observe the widgets
        self._slider.observe(on_slider_value_change, names='value')
        self._toggle_buttons.observe(on_toggle_button_change, names='value')

        # Initial display of all controls and the first image
        with self.output_area:
            clear_output() # Clear any previous run output
            display(HTML('<b>Final Output View:</b>'))
            display(self._slider)
            display(self._toggle_buttons)
            display(self._image_display_output) # Initially empty, will be filled by the next call
            display(self.save_button)

        # Manually trigger the first display with the initial slider value and toggle value
        on_slider_value_change({'new': self._slider.value})
````
