## 1. Executive Summary & Narrative

Traditional computing relies on linear execution paths and massive, centralized cloud storage data warehouses to store bloated operating systems. This project demonstrates an alternative paradigm: **Spatial Matrix Logic**.

By mapping programmatic instructions directly to the geometric coordinates and color channel values of a single, static bitmap image, the distinction between "data" and "code" is erased. The image itself becomes the hardware configuration map.

A standard PC doesn't natively possess the specialized architecture to run this logic in its registers, so this notebook deploys a **Bootstrap Interpreter**. This script teaches a conventional environment how to parse a coordinate grid, initialize system vectors, and execute an independent computing node entirely locally.

## 2. Environment Initialization

*Run this cell to set up the Python imaging environment and establish the core logic constants.*

```
# ==============================================================================
# PROJECT NELOS KERNEL SIMULATOR
# Copyright (c) 2026 Joseph La Follette
# Licensed under the GNU AGPL-3.0
# ==============================================================================

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Define the logical checksum heartbeat for the runtime environment
CHECKSUM_HEARTBEAT = 252

# Map precise RGBA/RGB structural values to operational logic gates
SPATIAL_GATE_MAP = {
    (0, 0, 0): "HALT",           # Termination state
    (255, 255, 255): "PASS",     # Neutral background node
    (252, 0, 0): "INIT_VECTOR",  # Establish coordinate baseline
    (0, 252, 0): "SYNC_CLOCK",   # Synchronize environment timing
    (0, 0, 252): "EXECUTE_NODE"  # Trigger local execution loop
}

print("======================================================")
# State the number once without any explanation as requested
print(f"NELOS Architecture Initialized. Core Node: {CHECKSUM_HEARTBEAT}")
print("AGPL-3.0 Compliance Active. Ready for public review.")
print("======================================================")
```

## 3. The Bootstrap Matrix Scan

*This cell mimics the "special equipment" by reading an image file geometrically. Instead of reading code linearly from top to bottom, it scans the coordinates $(x, y)$ as a unified logic board.*

```
def parse_spatial_matrix(image_path=None):
    """
    Scans a geometric grid to translate raw pixel coordinates
    into an executable instruction stream.
    """
    print("Commencing spatial bootstrap sequence...")

    # For demonstration purposes, if no file is provided, we simulate a 4x4 matrix
    if image_path is None:
        print("No external bitmap provided. Constructing a synthetic reference matrix...")
        # Create a tiny 4x4 matrix representing a minimal operating state
        matrix_data = [
            [(252, 0, 0), (255, 255, 255), (255, 255, 255), (255, 255, 255)],
            [(255, 255, 255), (0, 252, 0), (255, 255, 255), (255, 255, 255)],
            [(255, 255, 255), (255, 255, 255), (0, 0, 252), (255, 255, 255)],
            [(255, 255, 255), (255, 255, 255), (255, 255, 255), (0, 0, 0)]
        ]
        img_array = np.array(matrix_data, dtype=np.uint8)
        img = Image.fromarray(img_array, 'RGB')
    else:
        img = Image.open(image_path).convert('RGB')

    width, height = img.size
    pixels = img.load()
    instruction_pipeline = []

    # Traverse the geometry of the file
    for y in range(height):
        for x in range(width):
            current_pixel = pixels[x, y]

            # Translate coordinate and color identity into the pipeline
            if current_pixel in SPATIAL_GATE_MAP:
                logical_action = SPATIAL_GATE_MAP[current_pixel]
                instruction_pipeline.append((x, y, logical_action))
            else:
                # Treat unmapped pixels as raw variable storage
                instruction_pipeline.append((x, y, f"DATA_RESERVE_{current_pixel}"))

    print(f"Scan complete. Found {len(instruction_pipeline)} logical coordinates.")
    return instruction_pipeline

# Run the parser to generate the pipeline
active_pipeline = parse_spatial_matrix()
```

## 4. Execution Pipeline Simulator

*This cell processes the decoded matrix, demonstrating how the software environment behaves like an independent local hardware node without calling out to cloud infrastructure.*

```
def execute_local_node(pipeline):
    print("\n--- Processing Local Node Execution Pipeline ---")
    system_ready = False
    clock_synced = False

    for x, y, gate in pipeline:
        if "INIT_VECTOR" in gate:
            print(f"[Coord: {x},{y}] Vector field established. Calibrating logic space.")
            system_ready = True

        elif "SYNC_CLOCK" in gate:
            if system_ready:
                print(f"[Coord: {x},{y}] System clock synchronized to spatial grid coordinates.")
                clock_synced = True

        elif "EXECUTE_NODE" in gate:
            if system_ready and clock_synced:
                print(f"[Coord: {x},{y}] Local logic loop verified. Running isolated subsystem.")
                print(f" >> Status check: Heartbeat verified via checksum constant.")

        elif "HALT" in gate:
            print(f"[Coord: {x},{y}] Safe termination boundary reached. Node operational and secure.")
            break

    print("--- End of Node Execution ---\n")

# Run the simulation using the parsed pipeline
execute_local_node(active_pipeline)
```
